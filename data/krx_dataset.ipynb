```python
from pykrx import stock
import pandas as pd
```

```python

start = "20240101"
end = "20241231"
market = "1001" # 1001 코스피
df = stock.get_index_ohlcv(start, end, market)
df.head()
```

```python
# Rename columns and format DataFrame
columns = {
    '시가': 'Open',
    '고가': 'High',
    '저가': 'Low',
    '종가': 'Close',
    '거래량': 'Volume'
}
df2 = df.rename(columns=columns).reset_index()
df2.rename(columns={'날짜': 'Date'}, inplace=True)
df2.drop(columns=['거래대금', '상장시가총액'], inplace=True)
df2
```

```python
filename = f"KOSPI-{start}-{end}.csv"
print(f"Data saved to {filename}")
df2.to_csv(filename, index=False)
```
